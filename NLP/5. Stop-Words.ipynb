# Stop Words


```python
# Perform standard imports:
import spacy
nlp = spacy.load('en_core_web_sm')
```

```python
# Print the set of spaCy's default stop words 
# (remember that sets are unordered):
print(nlp.Defaults.stop_words)
```

```python
len(nlp.Defaults.stop_words)
```

## To see if a word is a stop word

```python
nlp.vocab['dev'].is_stop
```

## To add a stop word


```python
# Add the word to the set of stop words. Use lowercase!
nlp.Defaults.stop_words.add('dev')
```

```python
# Set the stop_word tag on the lexeme
nlp.vocab['dev'].is_stop = True
```

```python
len(nlp.Defaults.stop_words)
```

```python
nlp.vocab['get'].is_stop
```

## To remove a stop word


```python
# Remove the word from the set of stop words
nlp.Defaults.stop_words.remove('beyond')
```

```python
# Remove the stop_word tag from the lexeme
nlp.vocab['beyond'].is_stop = False
```

```python
len(nlp.Defaults.stop_words)
```

```python
nlp.vocab['beyond'].is_stop
```

```python
a = "hi my name Is dev And I live In andheri and i work in barclays"
len(set(a.lower().split()))
```

```python
stop  = list(nlp.Defaults.stop_words)

ans = []
for i in a.lower().split():
    if i not in stop:
        ans.append(i)
" ".join(ans)
```
