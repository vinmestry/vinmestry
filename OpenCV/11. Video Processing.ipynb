### Rotate Video

```python
def rotate(image, angel, center=None, scale=1.0):
    
    (h,w) = image.shape[:2]
    
    if center==None:
        center = (w//2, h//2)
        
    # create matrix for rotation
    M = cv2.getRotationMatrix2D(center, angel, scale)
    rotated = cv2.warpAffine(image, M, (w, h))
    
    return rotated
```

```python
# cap = cv2.VideoCapture(0)  
cap = cv2.VideoCapture("DATA/exp8.mp4")
 

time.sleep(2)
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
print(width, height)
 
angle = 0

while True:
    
    
    if angle==360:
        angle = 0
    
    # Capture frame-by-frame
    ret, frame = cap.read()
    # --------------------------------------------
    print("ret", ret)
    
    if ret == False:
        break
        
    frame = cv2.resize(frame, None, fx=0.7, fy=0.7)
    
    rframe = rotate(frame, angle)
      
    angle += 1
    time.sleep(0.06)
    # Display the resulting frame
#     cv2.imshow('frame',frame)
    cv2.imshow('rotated frame',rframe)
    # ------------------------------------------
    
    # This command let's us quit with the "q" button on a keyboard.
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
```

### Coneecting to mobile camera

```python
# Connects to your computer's default camera
# cap = cv2.VideoCapture(0)
cap = cv2.VideoCapture("DATA/exp8.mp4")

width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

while True:
    
    # Capture frame-by-frame
    ret, frame = cap.read()
    
    if ret == False:
        break
        
    img = cv2.resize(frame, None, fx=0.5, fy=0.5)
    
    (B, G, R) = cv2.split(img)

    z = np.zeros(img.shape[:2], dtype="uint8")
    
#     cv2.imshow("Red" , R)
#     cv2.imshow("Green" , G)
#     cv2.imshow("Blue", B)

    cv2.imshow("Red1" , cv2.merge([z, z, R]))
    cv2.imshow("Green1" , cv2.merge([z, G, z]))
    cv2.imshow("Blue1", cv2.merge([B,z,z]))

    # Display the resulting frame
    cv2.imshow('orignal',img)
    
    # This command let's us quit with the "q" button on a keyboard.
    # Simply pressing X on the window won't work!
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# When everything done, release the capture and destroy the windows
cap.release()
cv2.destroyAllWindows()
```

## Writing a Video Stream to File

### Notebook Users:  Make sure its all in the same cell!

FourCC is a 4-byte code used to specify the video codec. The list of available codes can be found in fourcc.org. It is platform dependent. 

MORE INFO ON CODECS: https://docs.opencv.org/3.0-beta/doc/py_tutorials/py_gui/py_video_display/py_video_display.html#saving-a-video

```python
import cv2

# cap = cv2.VideoCapture(0)
cap = cv2.VideoCapture("DATA/vid2.mp4")

time.sleep(2)

# Automatically grab width and height from video feed
# (returns float which we need to convert to integer for later on!)
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
print("w-{} h-{} f-{}".format(width,height,frames))


# MACOS AND LINUX: *'XVID' (MacOS users may want to try VIDX as well just in case)
# WINDOWS *'VIDX'
writer = cv2.VideoWriter('DATA/my_capture5.mp4', cv2.VideoWriter_fourcc(*'VIDX'), 15, 
                         (width, height))


## This loop keeps recording until you hit Q or escape the window
## You may want to instead use some sort of timer, like from time import sleep and then just record for 5 seconds.

while True:
    # Capture frame-by-frame
    ret, frame = cap.read()
    
    if ret == False:
        break
        
    ##################
    img = cv2.resize(frame, (960, 540))
    
    (B, G, R) = cv2.split(img)

    z = np.zeros(img.shape[:2], dtype="uint8")
    
    red_ch =   cv2.merge([z, z, R])

    cv2.imshow("Red" ,red_ch)
    
    #################
    
    # Write the video
    writer.write(red_ch)

    # Display the resulting frame
    cv2.imshow('frame',red_ch)
    
    # This command let's us quit with the "q" button on a keyboard.
    # Simply pressing X on the window won't work!
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

        
cap.release()
writer.release()
cv2.destroyAllWindows()
```
